```python
# Import Libraries 
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import scipy
```

Making a Basic figure :

```python
#Make sample data
np.random.seed(613) # set random seed to make it reproducable 
x=np.arange(50)
y=np.random.randint(0,100,50) # fifty random numbers betwwen 0 and 100

# make basic scaterplot
fig, ax=plt.subplots(figsize=(5,3)) #fig is our container 
ax.scatter(x,y)
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.bar(x,y) # This time we want to use bar plot with the same X and Y variable s before
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y) # line plot 
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.hist(y) #  produce a histogram
```

Plot Labels :

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('years since start')
```

We can also use font dictionaries to modify titles and labels:

```python
# make dictionaries with properites we want for our labels
font1={'family':'sans-serif','color':'blue','size':20} 
font2={'family':'monospace','color':'green','size':14}

fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y)

# now we use our font dictionaroes when seting our labels and titles 
ax.set_title('Total growth over time',fontdict=font1)
ax.set_ylabel('Total Growth',fontdict=font2)
ax.set_xlabel('years since start',fontdict=font2)
```

Customizing plot appearance :

matplotlib markers documentation : 
https://matplotlib.org/stable/api/markers_api.html

```python
# modify data points
fig, ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y,
           marker='*',
           color='indigo')
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2)
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2,
           markeredgecolor='#FA9359',
           markerfacecolor='#000000')
```

```python
#add grid
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2,
           markeredgecolor='#FA9359',
           markerfacecolor='#000000')
ax.grid() # we can add grid lines both horizontal and vertical 
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2,
           markeredgecolor='#FA9359',
           markerfacecolor='#000000')
ax.grid(axis = 'y') # horizontal grid line
```

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2,
           markeredgecolor='#FA9359',
           markerfacecolor='#000000')
ax.grid(axis = 'x' , color= 'dodgerblue') # vertical grid line and add color to that
```
