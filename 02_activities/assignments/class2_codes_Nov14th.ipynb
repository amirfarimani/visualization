```python
# loading libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy 

# settiing random seed for reproducabilibty
np.random.seed(613)
x=np.arange(50)
# this time we have two y variables 
y1=np.random.randint(0,100,50)
y2=np.random.randint(0,100,50)
```

Making a line plot with two y variables on the same axes:

```python
fig, ax=plt.subplots(figsize=(5,3)) 
ax.plot(x,y1)
ax.plot(x,y2)
```

Adding a legend :

```python
fig, ax=plt.subplots(figsize=(5,3))
# add a "label" argument to each plot
ax.plot(x,y1,label="Person 1")
ax.plot(x,y2,label="Person 2")
# call "legend" and specify legend location
ax.legend(loc="lower right", 
          frameon=True, # frame around the legend
          fontsize=12, # font size
          ncol=2, # number of colummns
          shadow=True) # shadow arround the frame
```

Moving the legend outside of the plot area :

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label="Person 1")
ax.plot(x,y2,label="Person 2")
ax.legend(loc= "upper left",
          bbox_to_anchor=(1,1)) # decide where we want the anchor point to be
```

Text and Annotations :

```python
# put data into a scatter plot.
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")

#use text() function to add annotation to specify coordinates and text
ax.text(10,95,"This valus is important!")
```

```python

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")
ax.text(10,95,"This valus is important!",
        ha="center", # for horizontal alignment
        color="red", # we could also use hex code
        size=20)
```

```python

fig, ax = plt.subplots()
ax.axis([0,10,0,10]) # create axis 

ax.text(1,5,".Data:(1,5)",transform=ax.transData) # the 1,5 refrences the data on axes 
ax.text(0.5,0.1,".Axes:(0.5,0.1)",transform=ax.transAxes) # refrencing axes cordinates
ax.text(0.2,0.2,".Figure:(0.2,0.2)",transform=fig.transFigure) # refrencing figure cordinates
```

Annotating with arrows:

```python
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")

ax.annotate("This is important!", xy=(10,95), #the point we want our arrow to point at
            xytext=(20,94), # the point we want our text to go
            arrowprops=dict(facecolor = 'black', shrink=0.05)) # properties of our arrow
```

```python
fig, ax =plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")

ax.annotate("This is important!", xy=(10,95), #the point we want our arrow to point at
            xytext=(20,94), # the point we want our text to go
            arrowprops=dict(facecolor = 'black',arrowstyle="wedge")) # this time with a different arrow style
```

Axis labels

Removing tick marks and labels :

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc='lower right')

ax.yaxis.set_major_locator(plt.NullLocator()) #locator to remove ticks and laberls
ax.xaxis.set_major_formatter(plt.NullFormatter())  # fromatter to remove only labels
```

Define max amount of ticks we want on our axis :

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) # limit to 3 ticks
```

Set tickmarks at particular intervals :

```python
fig, ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) # tickmakrs every five units
```

Rotate axis labels :

```python
# we use xticks() function to rorate axis labesland horizontal alignment
fig, ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="Person 1")
ax.scatter(x,y2,label="Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation=45,
           ha="right",
           fontsize=12,
           font='Serif')
```

Modify axis title :


```python
font1 = {'family':'serif','color':'indigo'} #defining font library
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label = "Person 1")
ax.scatter(x,y2,label = "Person 2")
ax.legend(loc = "lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation = 45,
           ha = "right",
           fontsize = 12,
           font='Serif')

plt.xlabel('Shiny New X Asis!',fontsize = 18, fontdict = font1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1=np.arange(50)
y1=np.random.randint(0, 75, 50)
x2=np.array(["Luffy","Zoro","Nami", "Usopp", "Sanji"]) # some categorical data
y2=np.array([110,180,240,99,220]) # an array of specific values
```

Now we want to have two plots next to each other, so we just have to define multiple axes and their relative positions.

```python
fig, (ax1, ax2)= plt.subplots(ncols=2, # number of columns 
                             nrows=1,# number of rows
                             figsize=(7,3)) 
ax1.scatter(x1,y1)
ax2.bar(x2,y2)



```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,nrows=1,figsize=(9,3))

font1={'family' : 'serif','size':10}
font2={'family' : 'serif','size':14}

ax1.scatter(x1,y1,marker='o',color='red')
ax2.bar(x2,y2,color='darkgreen')

ax1.set_title('Importtant Scatterplot',fontdict=font2)
ax1.set_ylabel('Measure',fontdict=font1)
ax1.set_xlabel('Time',fontdict=font1)

```

```python
fig,someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize=(7,4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)


```

Adding very large x axis titles :

```python
fig,someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize=(7,4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furoius", fontsize= 18)
         
```

Use contrainted layout  to make sure all our elements fit in :

```python
fig,someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                   ['ax2','ax3']],
                                   figsize=(7,4),
                                   layout="constrained") # add this argument

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furoius", fontsize= 18)
```
